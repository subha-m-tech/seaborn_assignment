# Univariate Analysis

```python
import seaborn as sns
```

```python
sns.distplot(dataset['ssc_p'], hist=True, kde=True, kde_kws= {'color': 'blue'}, color='green') 
# hist shows the details of repetition or frequency
```

```python
import matplotlib.pyplot as plt
```

# Bivariate Analysis

```python
sns.jointplot(x='ssc_p', y= 'salary', data=dataset) # use join plot if you want to see histrogram and scatter plot
plt.show()
```

```python
# Use join plot if you want to see histrogram and scatter plot. if you want hexagon kind of scatter plots use hex in kind
sns.jointplot(x='ssc_p', y= 'salary', data=dataset, kind='hex')
plt.show()
```

```python
# use join plot if you want to see histrogram and scatter plot. if you want kernel density 
# estimation kind of scatter plots use kde in kind
sns.jointplot(x='ssc_p', y= 'salary', data=dataset, kind='kde') 
plt.show()
```

```python
#hue is for differentiation, diag_kind -  diagnoal, palette - color combination husl
sns.pairplot(dataset, hue='gender', diag_kind='hist', kind= 'scatter',palette= 'husl')
plt.show()
```

# Categorical with numerical plots

```python
# if you want to calculate how much 10th marks students got in each 10th board
# in this case use stripplot
sns.stripplot(x='ssc_b', y= 'ssc_p', data=dataset)
plt.show()
```

```python
#Jitter in a strip plot is like adding a little randomness to each data point's position along the x-axis, 
#preventing them from overlapping too much and helping to see the distribution clearly
sns.stripplot(x='ssc_b', y= 'ssc_p', data=dataset, jitter=True)
plt.show()
```

```python
# A swarm plot is a type of scatter plot that is used for representing categorical values. 
# It is very similar to the strip plot, but it avoids the overlapping of points. We can use the seaborn.swarmplot() to create such graphs.
# It is not advisable to use this type of graph when the sample size is large
sns.swarmplot(x='ssc_b', y= 'ssc_p', data=dataset)
plt.show()
```
